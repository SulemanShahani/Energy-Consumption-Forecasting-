```python
# Function to create time series features
def create_features_and_lags(df, fourier_terms=4):



    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.day_of_year
    df['dayofmonth'] = df.index.day
    # Calculate the week of the year To calculate the week of the year, we can use the strftime method with the %U format code, which represents the week number of the year (Sunday as the first day of the week)
    df['weekofyear'] = df.index.strftime('%U').astype(int)






    # Lagged variables to capture long-term seasonality
    target_map = df['PJME_MW'].to_dict()
    df['lag1_year'] = (df.index - pd.Timedelta('364 days')).map(target_map)
    df['lag2_year'] = (df.index - pd.Timedelta('728 days')).map(target_map)
    df['lag3_year'] = (df.index - pd.Timedelta('1092 days')).map(target_map)







    # Fourier terms
    time = np.arange(len(df))
    for i in range(1, fourier_terms + 1):
        df[f'fourier_sin_{i}'] = np.sin(2 * np.pi * i * time / len(df))
        df[f'fourier_cos_{i}'] = np.cos(2 * np.pi * i * time / len(df))



    return df

df = create_features_and_lags(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX
import holidays
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Assuming create_features_and_lags is a function you defined earlier

# Time Series Cross-Validation (Forward Chaining)
tscv = TimeSeriesSplit(n_splits=5, test_size=24*365*1, gap=24)
df = df.sort_index()

# Lists to store results
fold = 0
cv_scores = []
predictions = []
true_values = []
dates = []

for train_index, test_index in tscv.split(df):
    train = df.iloc[train_index]
    test = df.iloc[test_index]

    # Create features for train and test sets
    train = create_features_and_lags(train)
    test = create_features_and_lags(test)

    FEATURES = ['dayofyear', 'hour', 'dayofweek', 'quarter', 'month', 'year'] + \
               [f'fourier_sin_{i}' for i in range(1, 5)] + \
               [f'fourier_cos_{i}' for i in range(1, 5)]
    TARGET = 'PJME_MW'

    X_train = train[FEATURES]
    y_train = train[TARGET]

    X_test = test[FEATURES]
    y_test = test[TARGET]

    # Initialize scaler
    scaler = StandardScaler()

    # Fit and transform on training data
    X_train_scaled = scaler.fit_transform(X_train)

    # Transform test data (using the same scaler)
    X_test_scaled = scaler.transform(X_test)

    # Initialize XGBoost model
    reg = xgb.XGBRegressor(
        base_score=0.5,
        booster='gbtree',
        n_estimators=1000,
        early_stopping_rounds=50,
        objective='reg:linear',
        max_depth=3,
        learning_rate=0.01
    )

    reg.fit(X_train_scaled, y_train,
            eval_set=[(X_train_scaled, y_train), (X_test_scaled, y_test)],
            verbose=100)

    # Make predictions
    y_pred = reg.predict(X_test_scaled)
    predictions.extend(y_pred)
    true_values.extend(y_test)
    dates.extend(test.index)

    # Evaluate RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    cv_scores.append(rmse)

    fold += 1
    print(f'Fold {fold}, RMSE: {rmse:.2f}')
    print(f'Fold {fold}, len(y_test): {len(y_test)}, len(y_pred): {len(y_pred)}')

# Print average RMSE across folds
print(f'Average RMSE across {tscv.n_splits} folds: {np.mean(cv_scores):.2f}')
print(f'Total lengths - dates: {len(dates)}, true_values: {len(true_values)}, predictions: {len(predictions)}')


```

```python
# Convert predictions and true values to a DataFrame
results_df = pd.DataFrame({
    'Date': dates,
    'True Values': true_values,
    'Predictions': predictions
}).set_index('Date')

# Plot predictions vs true values
plt.figure(figsize=(15, 5))
plt.plot(results_df['True Values'], label='True Values', color='b')
plt.plot(results_df['Predictions'], label='Predictions', color='r', linestyle='--')
plt.legend()
plt.title('True Values vs Predictions')
plt.xlabel('Time')
plt.ylabel('Energy Consumption (PJME_MW)')
plt.show()
```
